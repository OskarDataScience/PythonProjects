```python
# This is a small sheet with vector fields that I wrote for some of my students in order to improve on their intuition
# The idea is to have a methodical review and start from simple functions and move trough more complex functions
# Also, since the number of examples is never big enough there is always room for additional examples to improve the sheet
```

```python
# Vector fields are assignments of vectors in each point in a subset of space 
# They are functions in the form of F=Vx(x,y)i + Vy(x,y)j , where Vx and Vy are functions of both x and y coordinates
# and i and j are unit vectors representing the directions
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

%reset -f
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
```

```python
# To begin an example of a simple vector field only has +Vx component

x,y = np.meshgrid(np.linspace(-5,5,10), np.linspace(-5,5,10))

Vx = 1
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# This is an example where a vector field has only +Vy component

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = 0
Vy = 1

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# Now lets see how a vector field with Vx = Vy >0 looks like:

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = 1
Vy = 1

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# Now in the negative direction

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = -1
Vy = -1

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# Now lets see when we have a simple function Vx=x

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = x
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# The field is increasing as were moving away from the y-axis 
```

```python
# Now lets see how does it look like when we swich the components so that only Vy increases

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = 0
Vy = y

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# Now were going to take a look what happens when both components increase as Vx=x and Vy=y

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = x
Vy = y

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# If we were to take the same components from the previous example, but negative:

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = -x
Vy = -y

plt.quiver(x,y,Vx,Vy)
plt.show()

# We would get the same field, only in the opposite direction
```

```python
# If we want it to increase more in one direction like: 

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

Vx = 3*x
Vy = y

plt.quiver(x,y,Vx,Vy)
plt.show()

# We see that the field doesnt spread out evenly anymore
```

```python
# What would happen if one component would increase as a square:

x,y = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))

Vx = x**2
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# As we are moving away from the origin both to the left and right, the direction is only along the positive x-axis
# but the magnitude is increasing 
```

```python
# If now we also add the Vy component:

x,y = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))

Vx = x**2
Vy = y**2

plt.quiver(x,y,Vx,Vy)
plt.show()

# Its noticeable that the vectors are increasing in both directions
```

```python
# And now if we add negative signs to both components? 

x,y = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))

Vx = -x**2
Vy = -y**2

plt.quiver(x,y,Vx,Vy)
plt.show()

# The field is the same, only the direction changed, as expected
```

```python
# what do we add variable x to the Vy component and variable y to the Vx component? Does that make sense? 

x,y = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))

Vx = x + y
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# Yes it does, because if we increase both x and y only for the Vx component, that means only Vx will increase as we move 
# away from the origin, and as we move diagonally the vector components are shrinking due to one of them being negative
```

```python
# Now more functions can be introduced, like the exponential and logarithmic

x,y = np.meshgrid(np.linspace(-20,20,20),np.linspace(-20,20,20))

Vx = np.exp(x)
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# As expected, the function Vx is again incrieasig only in the positive x direction as every other would
```

```python
# If we were to graph Vy with an exponential function as well, again we can expect the magnitudes of the vectors to increase
# diagonally just as in the previous cases

x,y = np.meshgrid(np.linspace(-20,20,20),np.linspace(-20,20,20))

Vx = np.exp(x)
Vy = np.exp(y)

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# If we add additional variables we get

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

Vx = np.exp(x) + y
Vy = np.exp(y)

plt.quiver(x,y,Vx,Vy)
plt.show()

# A certain assymetry was introduced and thats due to the y variable added in the Vx component
```

```python
# Also, we cant forget the logarithmic function

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

Vx = np.log(x)
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# The log function increases slowly and so is the vector magnitude
```

```python
# Vector fields with trigonometric functions look like this

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

Vx = np.sin(x)
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# the magnitude of the vectors is increasing and decreasing periodically
```

```python
# In both directions

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

Vx = np.sin(x)
Vy = np.sin(y)

plt.quiver(x,y,Vx,Vy)
plt.show()

# Again, its diagonally symmetric, as every time
```

```python
# One more example

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

Vx = np.sin(x)
Vy = np.cos(y)

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# Now lets do a combination of exponential and trig functions

x,y = np.meshgrid(np.linspace(0,10,20),np.linspace(0,10,20))

Vx = (np.sin(x))*(np.exp(x))
Vy = 0

plt.quiver(x,y,Vx,Vy)
plt.show()

# Sine function oscilates from -1 to 1 and exponential function grows so the magnitude of the vector field is increasing and
# changing its direction
```

```python
# How do we create rotating vector fields? Any suggestions?

x,y = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))

Vx = (y/((x**2  +  y**2)**(1.5)))
Vy = (x/((x**2  +  y**2)**(1.5)))

plt.quiver(x,y,Vx,Vy)
plt.show()
```

```python
# So far we have only vector fields in 2D, now we are going to see 3D so we need to expand the function F(x,y) to 
# F(x,y,z) = Vx(x,y,z) + Vy(x,y,z) + Vz(x,y,z)

fig = plt.figure()
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

Vx = 1
Vy = 0
Vz = 0

ax.quiver(x, y, z, Vx, Vy, Vz, length=0.1, color = 'black')

plt.show()

# As expected, we get only Vx in the positive x direction but now it is contained within a volume, while in 2D case we could
# only observe the vectors in one plane, the xy plane. It is as if we have took only one slice of the volume

# Also, we can expect the same behaviour of the components in 3D as in 2D but maby its a bit harder to visualize
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

Vx = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
Vy = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
Vz = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, Vx, Vy, Vz, length=0.1, color = 'black')

plt.show()
```
